```python
import matplotlib.pyplot as plt
import spicelib

from common import figure_save

%matplotlib widget
```

```python
data = spicelib.RawRead("../cosim/simulations/pwm_dac_tb.raw")

data.get_trace_names()
```

```python
time = data.get_trace("time")
out = data.get_trace("v(out)")
```

```python
plt.figure(figsize=(10, 6))
plt.plot(time, out)
plt.xlabel("Time [s]")
plt.ylabel("Output Voltage [V]")
plt.title("PWM DAC Output")
plt.grid()
plt.show()

figure_save(plt.gcf(), 'pwm_dac_output', format='svg')
```
